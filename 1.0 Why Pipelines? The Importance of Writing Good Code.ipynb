## Writing Production-Ready Code


Pipelines can be applied to all steps of the CRISP-DM (Cross-Industry Standard Process for Data Mining). However, they are especially useful for making your code readable and clean at the deployment stage.

## Learning Outcomes:
- What are Pipelines and How to Use Them?
    - Classes in Python
    - Pipeline Buidling Blocks: Transformers and Estimators
    - Transformers and Estimators for Working with Different Data Types
    - Pipelines
    - Feature Unions
- How can Pipelines Help Improve My Code?
    - Writing Better Code: Test-Driven Development (TDD) and Writing the Failing Test
    - Pipelines in TDD: Passing the Failing Test
    - How are your Pipelines Working?
    - Long-Form Exercise
- Putting it into Practice
    - Mini-competition using pipelines and TDD.
